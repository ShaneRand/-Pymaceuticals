### Analysis
In this study, 249 mice who were identified with SCC tumors received treatment with a range of drug regimens. Over the course of 45 days, tumor development was observed and measured. The purpose of this study was to compare the performance of Pymaceuticalsâ€™ drug of interest, Capomulin, against the other treatment regimens.

The executive team has tasked you with generating all of the tables and figures needed for the technical report of the clinical study. They have also asked you for a top-level summary of the study results
There is a strong positive correlation weith the mouse weight and average tumor volume. The correlation coefficient is 0.8419. This tells me that for evey averager tumor decrease there was weight loss. We also note that there was even gender distribution from the pie chart. Lastly, we know the regressed line for mouse weight and tumor volume for Capomulin could model changes in tumor volume based on the mouse weight changes (r-squared was .7089).

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

sum_df = new_df.groupby(['Drug Regimen'])

sum_mean = sum_df['Tumor Volume (mm3)'].mean()
sum_med = sum_df['Tumor Volume (mm3)'].median()
sum_var = sum_df['Tumor Volume (mm3)'].var()
sum_std = sum_df['Tumor Volume (mm3)'].std()
sum_sem = sum_df['Tumor Volume (mm3)'].sem()

sum_df = pd.DataFrame({
                        "Mean Tumor Volume": sum_mean,
                        "Median Tumor Volume" : sum_med,
                        "Tumor Volume Variance" : sum_var,
                        "Tumor Volume Std. Dev." : sum_std,
                        "Tumor Volume Std. Err." : sum_sem
                        })
sum_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
sum_adv_df = new_df.groupby(['Drug Regimen']).agg({'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', 'sem']}) 
# Using the aggregation method, produce the same summary statistics in a single line
sum_adv_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
sel_treatments = new_df.loc[(new_df["Drug Regimen"] == "Capomulin") | 
                                    (new_df["Drug Regimen"] == "Ramicane") |
                                    (new_df["Drug Regimen"] == "Infubinol") |
                                    (new_df["Drug Regimen"] == "Ceftamin"),
                                    ["Mouse ID","Drug Regimen","Timepoint","Tumor Volume (mm3)" ] 
                        ] 
sel_treatments
# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = sel_treatments.groupby(["Drug Regimen","Mouse ID"])["Timepoint"].max().reset_index()
max_timepoint

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_volume_df= max_timepoint.merge(new_df[["Mouse ID","Timepoint","Tumor Volume (mm3)"]], 
                                                 on = ["Mouse ID", "Timepoint"], how = "left")
tumor_volume_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = [[] for i in range(4)]

for i in range(len(tumor_volume_df["Tumor Volume (mm3)"])):
    if tumor_volume_df["Drug Regimen"][i] == treatments [0]:
        tumor_vol[0].append(tumor_volume_df["Tumor Volume (mm3)"][i])
    elif tumor_volume_df["Drug Regimen"][i] == treatments [1]:
        tumor_vol[1].append(tumor_volume_df["Tumor Volume (mm3)"][i])
    elif tumor_volume_df["Drug Regimen"][i] == treatments [2]:
        tumor_vol[2].append(tumor_volume_df["Tumor Volume (mm3)"][i])
    elif tumor_volume_df["Drug Regimen"][i] == treatments [3]:
        tumor_vol[3].append(tumor_volume_df["Tumor Volume (mm3)"][i])

tumor_vol
```

```python
#Create empty list to fill for quartile calculations
quartiles = [[] for i in range (4)]
lowerq = [[] for i in range (4)]
upperq = [[] for i in range (4)]
iqr = [[] for i in range (4)]
lower_bound = [[] for i in range (4)]
upper_bound = [[] for i in range (4)]
```

```python
#create an empty dictionary for outliers
outliers_list = {}

#initialize index for quartile bound
bound_index = 0

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for index, treatment in enumerate (treatments):
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor = tumor_volume_df.loc[tumor_volume_df["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    quartiles [index] = final_tumor.quantile([.25,.5,.75])
    
    # Determine outliers using upper and lower bounds
    lowerq [index]= quartiles[index][0.25]
    upperq [index]= quartiles[index][0.75]
    iqr [index] = upperq[index]-lowerq[index]
    
    print(f"The lower quartile of {treatment} is: {round(lowerq[index],2)}")
    print(f"The upper quartile of {treatment} is: {round(upperq[index],2)}")
    print(f"The interquartile range of {treatment} is: {round(iqr[index],2)}")
    
    lower_bound [index] = lowerq[index] - (1.5*iqr[index])
    upper_bound [index] = upperq[index] + (1.5*iqr[index])
    
    print(f"Potential outliers for {treatment} could exist below {round(lower_bound[index],2)}")
    print(f"Potential outliers for {treatment} could exist above {round(upper_bound[index],2)}")
    print("-----------------------------------------------------")
    
while bound_index <= 3:
    for i in range(len(tumor_volume_df["Tumor Volume (mm3)"])):
        if (tumor_volume_df["Tumor Volume (mm3)"][i] < lower_bound [bound_index])\
        & (tumor_volume_df["Drug Regimen"][i] == treatments[bound_index]):
            outliers_list[tumor_volume_df["Drug Regimen"][i]] = tumor_volume_df["Tumor Volume (mm3)"][i]
        
        elif (tumor_volume_df["Tumor Volume (mm3)"][i] > upper_bound [bound_index])\
        & (tumor_volume_df["Drug Regimen"][i] == treatments[bound_index]):
            outliers_list[tumor_volume_df["Drug Regimen"][i]] = tumor_volume_df["Tumor Volume (mm3)"][i]  
            
    #increase bound index
    bound_index += 1

# Get the outliers for each Drug Regiment
for drug, outliers in outliers_list.items():
    print(f'The outlier(s) for {drug}: {round(outliers,2)}')
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np 
 
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
 
# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
 
# Combine the data into a single DataFrame
data_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
 
# Filter data for Capomulin regimen
capomulin_data = data_df[data_df['Drug Regimen'] == 'Capomulin']
 
# Calculate average tumor volume for each mouse
avg_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().reset_index()
 
# Merge average tumor volume with mouse metadata
merged_data = pd.merge(avg_tumor_volume, mouse_metadata[['Mouse ID', 'Weight (g)']], on='Mouse ID')
 
# Create the scatter plot
plt.figure(figsize=(10, 8))
scatter = plt.scatter(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'], 
                      c='blue', alpha=0.7, edgecolors='black')
 
# Add labels and title
plt.xlabel('Mouse Weight (g)', fontsize=14)
plt.ylabel('Average Tumor Volume (mmÂ³)', fontsize=14)
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen', fontsize=16)
 
# Add a trend line
slope, intercept, r_value, p_value, std_err = linregress(merged_data['Weight (g)'], merged_data['Tumor Volume (mm3)'])
line = slope * merged_data['Weight (g)'] + intercept
plt.plot(merged_data['Weight (g)'], line, color='red', linestyle='--', label=f'Trend line (RÂ² = {r_value**2:.3f})')
 
# Add legend
plt.legend(fontsize=12)
 
# Adjust layout and display the plot
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
 
# Print statistical summary
print("\nStatistical Summary:")
print(f"Correlation coefficient: {r_value:.3f}")
print(f"R-squared value: {r_value**2:.3f}")
print(f"P-value: {p_value:.3e}")
```

## Correlation and Regression

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np 
 
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
 
# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
 
# Combine the data into a single DataFrame
data_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
 
# Filter data for Capomulin regimen
capomulin_data = data_df[data_df['Drug Regimen'] == 'Capomulin']
 
# Calculate average tumor volume for each mouse
avg_mouse_data = capomulin_data.groupby('Mouse ID').agg({
    'Weight (g)': 'first',
    'Tumor Volume (mm3)': 'mean'
}).reset_index()
 
# Calculate the correlation coefficient and linear regression
correlation, p_value = st.pearsonr(avg_mouse_data['Weight (g)'], avg_mouse_data['Tumor Volume (mm3)'])
slope, intercept, r_value, p_value, std_err = linregress(avg_mouse_data['Weight (g)'], avg_mouse_data['Tumor Volume (mm3)'])
 
# Create the scatter plot
plt.figure(figsize=(10, 8))
scatter = plt.scatter(avg_mouse_data['Weight (g)'], avg_mouse_data['Tumor Volume (mm3)'], 
                      c='blue', alpha=0.7, edgecolors='black')
 
# Add labels and title
plt.xlabel('Mouse Weight (g)', fontsize=14)
plt.ylabel('Average Tumor Volume (mmÂ³)', fontsize=14)
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen', fontsize=16)
 
# Add trend line
x_range = np.linspace(avg_mouse_data['Weight (g)'].min(), avg_mouse_data['Weight (g)'].max(), 100)
y_range = slope * x_range + intercept
plt.plot(x_range, y_range, color='red', linestyle='--', 
         label=f'Trend line: y = {slope:.2f}x + {intercept:.2f}')
 
# Add legend
plt.legend(fontsize=12)
 
# Adjust layout and display the plot
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
 
# Print statistical summary
print("\nStatistical Summary:")
print(f"Correlation coefficient: {correlation:.4f}")
print(f"R-squared value: {r_value**2:.4f}")
print(f"P-value: {p_value:.4e}")
print(f"\nLinear Regression Equation: y = {slope:.4f}x + {intercept:.4f}")
print(f"Standard Error: {std_err:.4f}")
```
